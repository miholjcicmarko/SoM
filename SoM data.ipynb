```python
pif_id = data["PIFID"]
pif_date = data["PIFDATE"]
pif_city = data["City"]
pif_st = data["State"]
data = data.drop(["PIFID", "PIFDATE", "City", "State"], axis = 1)
data["PIFID"] = pif_id
data["PIFDATE"] = pif_date
data["City"] = pif_city
data["State"] = pif_st
data
```

```python
for i in range(len(data.columns)):
    if data.dtypes[i] == "object" and i != 0:
        data = data.rename(columns={data.columns[i]: data.columns[i]+"Categ"})
    elif i == 0:
        data = data.rename(columns={data.columns[i]: "ID"})
data
```

```python
for i in range(len(data)):
    data["PIFDATECateg"][i] = "Date" + data["PIFDATECateg"][i][:8]
    data["PIFDATECateg"][i] = data["PIFDATECateg"][i].replace("/", "-")
    data["PIFIDCateg"][i] = data["PIFIDCateg"][i].replace(".", "dot")
    data["PIFIDCateg"][i] = data["PIFIDCateg"][i].replace("@", "at")
data
```

```python
data['ID'] = 'id' + data['ID'].astype(str)
```
